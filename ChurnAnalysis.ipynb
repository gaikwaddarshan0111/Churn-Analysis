```python
#converting 0 & 1 value of senior citizen to yes & No
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn' , data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of churnout cusomers")
plt.show()
```

```python
df.columns.values
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define the number of rows and columns for the subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols

# Set up the figure
plt.figure(figsize=(9,12))
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

# Loop through each column and plot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], palette='Set2', hue="Churn")
    axes[i].set_title(f'Count of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=35)

# Hide any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Above charts shows the churn out counts depending upon the various category 
We Can note following points:
- People using interner service are churning out more
- Same where we have more tech support involved they are likely to churn
